```python
import random
import matplotlib.pyplot as plt
def get_previousdata(classnum,sample_num_class_train,sample_num_class_test,randompick=False):  ## etc: classnum=(0,2,3)
  train_datapool=[ img_train[index] for index in classnum ]
  test_datapool=[ img_test[index] for index in classnum ]
  train_datagroup = []
  train_randompick_pool=[]
  train_x_randompick_pool = []
  train_y_randompick_pool = []
  test_datagroup = []
  test_randompick_pool=[]
  test_x_randompick_pool = []
  test_y_randompick_pool = []
  for i in range(len(classnum)):
    train_datagroup.append([])
    for j in range(len(train_datapool[i])):
      temp=[train_datapool[i][j]]
      temp.insert(28,classnum[i])
      train_datagroup[i].append(temp)

    train_classdata = random.sample(train_datagroup[i],sample_num_class_train)
    train_randompick_pool += train_classdata


  for i in range(len(classnum)):
    test_datagroup.append([])
    for j in range(len(test_datapool[i])):
      temp=[test_datapool[i][j]]
      temp.insert(28,classnum[i])
      test_datagroup[i].append(temp)

    test_classdata = random.sample(test_datagroup[i],sample_num_class_test)
    #test_classdata = test_datagroup[i][0:sample_num_class_test]
    test_randompick_pool += test_classdata


  random.shuffle(train_randompick_pool)
  random.shuffle(test_randompick_pool)
  for i in range(len(train_randompick_pool)):
    train_x_randompick_pool.append(train_randompick_pool[i][0])
    train_y_randompick_pool.append(train_randompick_pool[i][1])

  for i in range(len(test_randompick_pool)):
    test_x_randompick_pool.append(test_randompick_pool[i][0])
    test_y_randompick_pool.append(test_randompick_pool[i][1])


  return train_x_randompick_pool, train_y_randompick_pool, test_x_randompick_pool, test_y_randompick_pool,train_randompick_pool,test_randompick_pool


def get_newdata(newclass,train_randompick_pool,test_randompick_pool,sample_num_class_train,sample_num_class_test):
  train_datapool=img_train[newclass]
  test_datapool=img_test[newclass]
  train_datagroup = []
  train_x_randompick_pool = []
  train_y_randompick_pool = []
  test_datagroup = []
  test_x_randompick_pool = []
  test_y_randompick_pool = []
  for i in range(len(train_datapool)):
    temp=[train_datapool[i]]
    temp.insert(28,newclass)
    train_datagroup.append(temp)
  train_classdata = random.sample(train_datagroup,sample_num_class_train)
  train_randompick_pool_new =train_randompick_pool + train_classdata

  for i in range(len(test_datapool)):
    temp=[test_datapool[i]]
    temp.insert(28,newclass)
    test_datagroup.append(temp)
  test_classdata = random.sample(test_datagroup,sample_num_class_test)
  #test_classdata = test_datagroup[0:sample_num_class_test]
  test_randompick_pool_new=test_randompick_pool + test_classdata

  random.shuffle(train_randompick_pool_new)
  random.shuffle(test_randompick_pool_new)

  for i in range(len(train_randompick_pool_new)):
    train_x_randompick_pool.append(train_randompick_pool_new[i][0])
    train_y_randompick_pool.append(train_randompick_pool_new[i][1])

  for i in range(len(test_randompick_pool_new)):
    test_x_randompick_pool.append(test_randompick_pool_new[i][0])
    test_y_randompick_pool.append(test_randompick_pool_new[i][1])

  return train_x_randompick_pool, train_y_randompick_pool, test_x_randompick_pool, test_y_randompick_pool,train_randompick_pool_new,test_randompick_pool_new
```

# Build Detector

```python
#The Detection_Training Part is used for training the detection network
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,Dropout,Flatten,Activation,BatchNormalization,AveragePooling2D,MaxPooling2D
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import load_model
#from keras.layers.normalization import
def generate_dtrainingdata(train_x_randompick_pool,train_y_randompick_pool,test_x_randompick_pool,test_y_randompick_pool,classnumber):
  train_y=np.arange(len(train_y_randompick_pool))
  test_y=np.arange(len(test_y_randompick_pool))
  for i in range(len(train_y_randompick_pool)):
    if (train_y_randompick_pool[i]==classnumber):
      train_y[i]=0
    else:
      train_y[i]=1

  for i in range(len(test_y_randompick_pool)):
    if (test_y_randompick_pool[i]==classnumber):
      test_y[i]=0
    else:
      test_y[i]=1

  train_y=to_categorical(train_y)
  test_y=to_categorical(test_y)
  train_x=np.array(train_x_randompick_pool)
  train_x=train_x.reshape(train_x.shape[0],train_x.shape[1],train_x.shape[2],1)/255.0
  test_x=np.array(test_x_randompick_pool)
  test_x=test_x.reshape(test_x.shape[0],test_x.shape[1],test_x.shape[2],1)/255.0

  return train_x,train_y,test_x,test_y
### each detector is a CNN model, but convolutional layers will not be trained.
def create_model(CNN_weight_path):
  chanDim = -1
  input_shape = (28, 28, 1)
  model_original = load_model('/content/gdrive/MyDrive/accretionary_learning_tf/99.4_CNN.h5')
  #model_original = load_model(CNN_weight_path)
  model = Sequential()
# Layer 1 Conv2D
  cnn_layer=Conv2D(filters = 32, kernel_size = 5, strides = 1, activation = 'relu', input_shape = input_shape)
  cnn_layer.trainable=False
  model.add(cnn_layer)
  cnn_layer2= Conv2D(filters = 32, kernel_size = 5, strides = 1, use_bias=False)
  cnn_layer2.trainable=False
  model.add(cnn_layer2)
  cnn_layer3=BatchNormalization()
  cnn_layer3.trainable=False
  model.add(cnn_layer3)
  model.add(Activation('relu'))
  model.add(MaxPooling2D(pool_size = 2, strides = 2))
  model.add(Dropout(0.25))


# Layer 3 Conv2D
  cnn_layer4= Conv2D(filters = 64, kernel_size = 3, strides = 1, activation = 'relu')
  cnn_layer4.trainable=False
  model.add(cnn_layer4)
  cnn_layer5= Conv2D(filters = 64, kernel_size = 3, strides = 1, use_bias=False)
  cnn_layer5.trainable=False
  model.add(cnn_layer5)
  cnn_layer6= BatchNormalization()
  cnn_layer6.trainable=False
  model.add(cnn_layer6)
# Layer 4 Pooling Layer
  model.add(Activation('relu'))
  model.add(MaxPooling2D(pool_size = 2, strides = 2))
  model.add(Dropout(0.25))
  model.add(Flatten())
  model.add(Dense(500, use_bias=False))
  model.add(BatchNormalization())
  model.add(Activation('relu'))

  model.add(Dense(300, use_bias=False))
  model.add(BatchNormalization())
  model.add(Activation('relu'))
  model.add(Dense(100, use_bias=False))
  model.add(BatchNormalization())
  model.add(Activation('relu'))
  model.add(Dropout(0.25))
  model.add(Dense(20, use_bias=False))
  model.add(BatchNormalization())
  model.add(Activation('relu'))

  model.add(Dense(2, activation='softmax'))
  model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
  model.layers[0].set_weights(model_original.layers[0].get_weights())
  model.layers[1].set_weights(model_original.layers[1].get_weights())
  model.layers[2].set_weights(model_original.layers[2].get_weights())
  model.layers[6].set_weights(model_original.layers[6].get_weights())
  model.layers[7].set_weights(model_original.layers[7].get_weights())
  model.layers[8].set_weights(model_original.layers[8].get_weights())
  return model

def train_model(model, dataX, dataY, epoch_count,classnum,path_detector):
  history = model.fit(dataX, dataY, batch_size=16, epochs=epoch_count, validation_split=0.05)
  #path='/content/gdrive/MyDrive/accretionary_learning_tf/detector_weights/'
  weightpath=path_detector+str(classnum)
  model.save_weights(weightpath)


def test_model(model, dataX, dataY,classnum,path_detector):
  #path='/content/gdrive/MyDrive/accretionary_learning_tf/detector_weights/'
  weightpath=path_detector+str(classnum)
  model.load_weights(weightpath).expect_partial()
  loss, acc = model.evaluate(dataX, dataY, verbose=2)
  return acc

def pre_train(train_x_randompick_pool,train_y_randompick_pool,test_x_randompick_pool,test_y_randompick_pool,classnumber,epoch_count,CNN_weight_path,path_detector):
  test_acc=[]
  for num in classnumber:
    train_x,train_y,test_x,test_y=generate_dtrainingdata(train_x_randompick_pool,train_y_randompick_pool,test_x_randompick_pool,test_y_randompick_pool,num)
    model=create_model(CNN_weight_path)
    train_model(model,train_x,train_y,epoch_count,num,path_detector)
    acc=test_model(model,test_x,test_y,num,path_detector)
    test_acc.append(acc)

  return test_acc,classnumber
def expand_detector(train_x_randompick_pool, train_y_randompick_pool, test_x_randompick_pool, test_y_randompick_pool,classnumber,epoch_count,previous_class,CNN_weight_path,path_detector):
  detector_list=[]
  for i in range(len(previous_class)):
    detector_list.append(previous_class[i])
  detector_list.append(classnumber)
  train_x,train_y,test_x,test_y=generate_dtrainingdata(train_x_randompick_pool,train_y_randompick_pool,test_x_randompick_pool,test_y_randompick_pool,classnumber)
  model=create_model(CNN_weight_path)
  train_model(model,train_x,train_y,epoch_count,classnumber,path_detector)
  acc=test_model(model,test_x,test_y,classnumber,path_detector)
  return acc,detector_list
```

# Decision_Part

```python
import tensorflow.keras as keras
from tensorflow.keras import backend as keras_backend
def data_prepare(detector_list,train_randompick_pool,test_randompick_pool,CNN_weight_path,path_detector):
  train_x_randompick_pool = []
  train_y_randompick_pool = []
  test_x_randompick_pool = []
  test_y_randompick_pool = []
  newtrainingset = []
  newtestingset = []
  random.shuffle(train_randompick_pool)
  random.shuffle(test_randompick_pool)

  for i in range(len(train_randompick_pool)):
    train_x_randompick_pool.append(train_randompick_pool[i][0])
    train_y_randompick_pool.append(train_randompick_pool[i][1])

  train_x=np.array(train_x_randompick_pool)
  train_x=train_x.reshape(train_x.shape[0],train_x.shape[1],train_x.shape[2],1)/255.0
  train_y=to_categorical(train_y_randompick_pool)
  train_y=np.array(train_y)
  for i in detector_list:
    model=create_model(CNN_weight_path)
    #path='/content/gdrive/MyDrive/accretionary_learning_tf/detector_weights/'
    weightpath=path_detector+str(i)
    model.load_weights(weightpath).expect_partial()
    y_pred=model.predict(train_x)
    newtrainingset.append(np.transpose(y_pred))

  newtrainingset=np.array(newtrainingset)
  newtrainingset=newtrainingset.reshape(len(detector_list)*2,newtrainingset.shape[2])
  newtrainingset=np.transpose(newtrainingset)




  for i in range(len(test_randompick_pool)):
    test_x_randompick_pool.append(test_randompick_pool[i][0])
    test_y_randompick_pool.append(test_randompick_pool[i][1])

  test_x=np.array(test_x_randompick_pool)
  test_x=test_x.reshape(test_x.shape[0],test_x.shape[1],test_x.shape[2],1)/255.0
  test_y=to_categorical(test_y_randompick_pool)
  test_y=np.array(test_y)
  for i in detector_list:
    model=create_model(CNN_weight_path)
    #path='/content/gdrive/MyDrive/accretionary_learning_tf/detector_weights/'
    weightpath=path_detector+str(i)
    model.load_weights(weightpath).expect_partial()
    y_pred=model.predict(test_x)
    newtestingset.append(np.transpose(y_pred))

  newtestingset=np.array(newtestingset)
  newtestingset=newtestingset.reshape(len(detector_list)*2,newtestingset.shape[2])
  newtestingset=np.transpose(newtestingset)


  return newtrainingset,newtestingset,train_y,test_y

#create_decision_network: create decision network in the bootstrap stage
def create_decision_network(length):
  num=length
  model = Sequential()
  model.add(Dense(num, input_dim=2*num, activation='softmax',kernel_initializer='glorot_uniform'))
  model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
  return model

def bootstrap_train_model(model, dataX, dataY, epoch_count,classnum,path_decision):
  history = model.fit(dataX, dataY, batch_size=4, epochs=epoch_count, validation_split=0.05)
  #path='/content/gdrive/MyDrive/accretionary_learning_tf/decision_weights/'
  weightpath=path_decision+str(len(classnum))
  model.save_weights(weightpath)
  weight=model.trainable_variables[0]
  bias=model.trainable_variables[1]
  weight=weight.numpy()
  bias=bias.numpy()
  return weight,bias

def network_expanding(newclass,weight,bias,previousclass):
  length=len(previousclass)+1
  model_new=create_decision_network(length) #build new decision network
  weight_new=model_new.trainable_variables[0]
  bias_new=model_new.trainable_variables[1]
  weight_new=weight_new.numpy()
  bias_new=bias_new.numpy()
  for i in range(weight.shape[0]):
    for j in range(weight.shape[1]):
      weight_new[i][j]=weight[i][j]## load weights from the previous decision network
  for k in range(len(bias)):
    bias_new[k]=bias[k]
  model_new.trainable_variables[0].assign(tf.convert_to_tensor(weight_new, dtype=tf.float32))
  model_new.trainable_variables[1].assign(tf.convert_to_tensor(bias_new, dtype=tf.float32))
  return model_new


########## Training process
def loss_function(pred_y, y):
  return keras_backend.mean(keras.losses.categorical_crossentropy(y, pred_y))
  #return keras.losses.categorical_crossentropy(y, pred_y)

def np_to_tensor(list_of_numpy_objs):
    return (tf.convert_to_tensor(obj) for obj in list_of_numpy_objs)


def compute_loss(model, x, y, loss_fn=loss_function):
    logits = model(x)
    mse = loss_fn(y, logits)
    return mse, logits


def compute_gradients(model, x, y, loss_fn=loss_function):
    with tf.GradientTape() as tape:
        loss, _ = compute_loss(model, x, y, loss_fn)
    return tape.gradient(loss, model.trainable_variables), loss


def apply_gradients(optimizer, gradients, variables):
    optimizer.apply_gradients(zip(gradients, variables))


def train_batch(x, y, model, optimizer,weight_prev,bias_prev):
    tensor_x, tensor_y = np_to_tensor((x, y))
    gradients, loss = compute_gradients(model, tensor_x, tensor_y)
    grad_weight=gradients[0]
    grad_weight=grad_weight.numpy()
    grad_bias=gradients[1]
    grad_bias=grad_bias.numpy()
    for i in range(weight_prev.shape[0]):
      for j in range(weight_prev.shape[1]):
        grad_weight[i][j]=0 # frozen the gradient update for the previous decision network weights

    for k in range(len(bias_prev)):
      grad_bias[k]=0
    gradients[0]=(tf.convert_to_tensor(grad_weight, dtype=tf.float32))
    gradients[1]=(tf.convert_to_tensor(grad_bias, dtype=tf.float32))
    apply_gradients(optimizer, gradients, model.trainable_variables)
    return gradients,loss

def train_decision_model(newclass,weight,bias,previousclass,trainingset_x,trainingset_y,epochs=2, lr=0.001):
    model= network_expanding(newclass,weight,bias,previousclass)
    optimizer = keras.optimizers.Adam(learning_rate=lr)
    #x=trainingset_x
    #y=trainingset_y
    losses=[]

    for epochs in range(epochs):
      total_loss = 0
      for i in range(int(len(trainingset_x)/4)):
        x=trainingset_x[i:i+4]
        y=trainingset_y[i:i+4]
        _,loss = train_batch(x, y, model, optimizer,weight,bias)
        total_loss += loss
        curr_loss = total_loss / (i + 1.0)
        losses.append(curr_loss)

       #loss_record=loss.numpy()
       #losses.append(loss_record)
    plt.plot(losses)
    plt.title('Loss Vs Time steps')
    plt.show()
    return model,losses
################### train the model

def bootstrap_train_test_decision(detector_list,train_randompick_pool,test_randompick_pool,epoch,CNN_weight_path,path_detector,path_decision):
  newtrainingset,newtestingset,train_y,test_y=data_prepare(detector_list,train_randompick_pool,test_randompick_pool,CNN_weight_path,path_detector)
  bootstrap_net=create_decision_network(len(detector_list))
  weight,bias=bootstrap_train_model(bootstrap_net,newtrainingset,train_y,epoch,detector_list,path_decision)
  score = bootstrap_net.evaluate(newtestingset, test_y, verbose=0)
  return weight,bias,score,bootstrap_net

def train_test_new_decision(newclass,weight,bias,previousclass,train_randompick_pool,test_randompick_pool,steps,CNN_weight_path,path_detector,path_decision):
  detector_list=[]
  for i in range(len(previousclass)):
    detector_list.append(previousclass[i])
  detector_list.append(newclass)
  newtrainingset,newtestingset,train_y,test_y=data_prepare(detector_list,train_randompick_pool,test_randompick_pool,CNN_weight_path,path_detector)
  model_new,loss=train_decision_model(newclass,weight,bias,previousclass,newtrainingset, train_y,epochs=steps, lr=0.001)
  #path='/content/gdrive/MyDrive/accretionary_learning_tf/decision_weights/'
  weightpath=path_decision+str(len(detector_list))
  model_new.save_weights(weightpath)
  weight_new=model_new.trainable_variables[0]
  bias_new=model_new.trainable_variables[1]
  weight_new=weight_new.numpy()
  bias_new=bias_new.numpy()
  score = model_new.evaluate(newtestingset, test_y, verbose=0)
  return weight_new,bias_new,score,model_new
```
